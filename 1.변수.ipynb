### 인덱싱(indexing)
- 무엇인가를 가리킨다는 의미(특정 요소를 가리키기)
### 슬라이싱(Slicing)
- 무엇인가를 잘라낸다는 의미(잘라내서 가져오기)
- 슬라이싱을 할 때 변수[start:end] start = 출력 포함/ end = 출력 미포함

- 인덱스 번호는 앞에서 세면 0 ~ 마지막수, 뒤에서 세면 처음수 ~ -1
- print문 사용시 속성값이 기본적으로 쉼표가 들어가면 공백출력
- 이 때 공백값을 출력하지 않으려면 소괄호에 sep ='변경값' 입력하기(sep 기본값은 공백)

```python
s4 = "My name is KM"
print(s4[8], s4[-9], s4[5], s4[-2], s4[-1])
print(s4[0:2], s4[-10:-6], s4[8:10], sep = '')
```

```python
day = "2023년 3월 3일의 날씨는 맑음입니다."

print("날짜 : " ,day[0:11])
print("날씨 : " ,day[-6:-4])
```

### 문자열 포맷코드
- 문자열 내부에 값 삽입



```python
#1. % 기호 사용
day = 24
month = 7

s = '오늘은 %d월 %d일 입니다.'%(month,day)
s
```

```python
#2. format 함수 사용
day = 24
month = 7

s = '오늘은 {}월 {}일 입니다.' .format(month, day)
s
```

```python
#3. f 문자열 사용
day = 24
month = 7

s = f'오늘은 {month}월 {day}일 입니다.' 
s
```

```python
x = 100
y = 200
sum2 = x + y
'%d와 %d의 합은 %d입니다.'%(x,y,sum2)
```

```python
x = 100
y = 200
sum2 = x + y
'{}와 {}의 합은 {}입니다' .format(x,y,sum2)
```
