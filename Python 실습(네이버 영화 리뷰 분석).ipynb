```python
with open('./movie_review.pkl', 'rb') as f : 
    data = pickle.load(f)
```

```python
#[[' 1', (1) or (0)], [' 2', (1) or (0)]....]
data
```

```python
# 1.    
len(data)
```

```python
# 2.  
for i in data :
    print(i[0])
```

```python
# 3.    
max_len = data[0][0]
max_len_size = len(data[0][0])
list_num = 0
while list_num < len(data) :
    if len(max_len) < len(data[list_num][0]) :
        max_len = data[list_num][0]
        max_len_size = len(data[list_num][0])
    list_num += 1
print(f"   : {max_len_size}")
print(f"  : {max_len}")
```

```python
# 4.    
min_len = max_len
min_len_size = max_len_size
list_num = 0
while list_num < len(data) :
    if len(min_len) > len(data[list_num][0]) :
        min_len = data[list_num][0]
        min_len_size = len(data[list_num][0])
    list_num += 1
print(f"   : {min_len_size}")
print(f"  : {min_len}")
```

```python
# 5.   
list_len_sum = 0
list_num = 0
while list_num < len(data) :
    list_len_sum += len(data[list_num][0])
    list_num += 1
print(f"    : {list_len_sum/len(data)}")
```

```python
# 6.  /  ( = 1,  =0)

positive_data = 0
denial_data = 0
list_num = 0

for i in data :
    if i[1] == 0 :
        denial_data += 1
    else :
        positive_data += 1
    list_num += 1
print(f"  : {positive_data}")
print(f"  : {denial_data}")
```

```python
# 7.     (,  )
text_sum = 0
search = input("   >> ")
for i in data :
    if search in i[0] :
        text_sum += 1
print(f" \'{search}\'     : {text_sum}")
```
