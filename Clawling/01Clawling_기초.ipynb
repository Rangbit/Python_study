### BeautifulSoup library
- html 문서에서 원하는 데이터를 추출하는 라이브러리

```python
from bs4 import BeautifulSoup as bs
```

```python
# bs(parsing 할 html 문서, 'parsing 방법')
# lxml : 빠르고 관대함 / 스트링 타입으로 받아온 데이터를 객채화 한다
soup = bs(res.text,'lxml')
soup
```

### 네이버 검색페이지에서 정보 가져오기
- 메뉴 글자 수집하기

```python
soup.select('a')
```

```python
naver_tap = soup.select('a.tab')
naver_tap
```

```python
naver_tap[0]
```

```python
# select_one : 요소중에서 첫번째 요소만 단수형으로 가져옴
# select : 전체를 리스트 형태로 가져옴
soup.select_one('a.tab')
```

```python
naver_tap[0].text
```
