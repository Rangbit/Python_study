### Melon Top100 수집하기

```python
import requests as re
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.melon.com/chart/index.htm'
re.get(url)
# 멜론페이지는 보안문제 때문에 파이썬 코드로 접근하는걸 막고있다.
```

- No.4xx error >> 접근하려는 페이지에서 F12(개발자모드) >> Network >>
- F5(Reload)>> Name(index / document type html)  >>
- Headers : User-Agent key:value copy >> h = { key : value }  >>
- re.get(url,headers = h) >>

```python
# 딕셔너리 { : }  형태로 만들어준다
h = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'}
```

```python
# bs 객체화
melon = re.get(url,headers = h)
melon
```

```python
melon.text
```

```python
melon_chart = bs(melon.text, 'lxml')
melon_chart
```

```python
## 노래제목, 가수 
rank1_music = melon_chart.select('div.wrap_song_info span')
rank1_music
```

```python
rank1_music[0]
```

```python
rank1 = rank1_music[0].text
rank1[1:10]
```

```python
rank1_music[1].text
```

```python
# 빈리스트가 출력된다
# 선택자 이름안에 공백은 자손선택자를 의미, 그래서 rank01을 자손이라고 생각함
# 해결방법 : 공백을 온점으로 대체
music = melon_chart.select('div.ellipsis.rank01>span>a')
music
```

```python
# 가수이름(아티스트) 수집하기
artist = melon_chart.select('div.ellipsis.rank02>span')
artist
```

```python
# 반복문 사용하기
# 1.빈 공간 생성(빈 리스트 생성)
# 2.컨텐츠 인덱스 번호 순서로 추출
# 3.추출한 컨텐츠를 빈 공간에 한 개씩 담기

musicList = []
artistList = []
for i in range(len(music)) : 
    musicList.append(music[i].text)
musicList
```

```python
for j in artist :
    artistList.append(j.text)
artistList
```

```python
for i in range(len(musicList)) : 
    print(f'{i+1}위. {musicList[i]} - {artistList[i]}')
```

```python
rankList = []
for i in range(len(music)) : 
    rankList.append(i+1)
rankList
```

```python
import pandas as pd # 데이터프레임 형식으로 만들기 위한 라이브러리
```

```python
dic = {'순위' : rankList, '노래제목' : musicList, '가수' : artistList}
dic
```

```python
melon = pd.DataFrame(dic)
```

```python
melon.set_index('순위', inplace = True)
```

```python
melon
```

```python
# DataFrame csv 파일로 꺼내보자
melon.to_csv('멜론차트.txt', encoding='euc-kr')
```
