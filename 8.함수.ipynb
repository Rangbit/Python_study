## 함수(function)
- 함수란?
    - 호출 될 때 특정 기능을 실행하는 재사용 가능한 코드블록
- 함수의 목적
    - 반복적인 프로그래밍을 피할 수 있다.
    - 모듈화(코드모음)로 인해 전체적인 코드의 가독성이 좋아진다.
    - 유지보수가 용이하다.

```python
# 함수 정의
def number_sum(num1, num2) : 
    result = num1 + num2
    return result
```

```python
n1 = 3
n2 = 10
print(number_sum(n1,n2))
```

```python
def number_sub(num1,num2) :
    if num1 > num2 :
        result = num1 - num2
    else : 
        result = num2 - num1
    return result
```

```python
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("첫 번째 정수 입력 >> "))
result = number_sub(num1,num2)
print(result)
```

### 독스트링(docstring)
- 함수의 기능설명 작성
- 정의된 함수 내에서 Shift + tab 누르면 설명 확인 가능
- 함수 내부 첫 줄에 개행을 포함한 긴 줄 문자열 기호 (""" 설명문 """)로 작성

```python
def cal(num1, num2, op) :
    """두 수를 입력받아서 원하는 연산을 수행하는 함수""" #독스트링
    if op == '+' :
        result = num1 + num2
    elif op == '-' :
        result = num1 - num2 if n1 > n2 else num2 - num1
    elif op == '*' :
        result = num1 * num2
    elif op == '/' :
        result = num1 / num2
    else :
        result = print("잘못 입력하셨습니다.")
    return result
```

```python
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("첫 번째 정수 입력 >> "))
op = input("연산자 입력(+,-,*,/)")

result = cal(num1, num2, op)
print(f"결과 : {result}")
```

### 함수 매개변수 - 가변 매개변수 (*)
- 함수 호출시 몇 개의 인수가 전달될지 알 수 없다면, 사용자가 직접 매개변수의 개수를 정할 수 있도록 선언

```python
#가변 매개변수로 함수를 정의
def add(*args) :
    result = sum(args)
    return result
```

```python
# 정해진 갯수가 없이 입력가능
add(8, 9, 8, 7, 4, 6)
```
